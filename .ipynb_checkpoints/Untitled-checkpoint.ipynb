```python
roles = [
{
"Id": 1,
"Name": "System Administrator",
"Parent": 0
},
{
"Id": 2,
"Name": "Location Manager",
"Parent": 1
    
},
{
"Id": 3,
"Name": "Supervisor",
"Parent": 2
},
{
"Id": 4,
"Name": "Employee",
"Parent": 3
},
{
"Id": 5,
"Name": "Trainer",
"Parent": 3
}
]
```

```python
users= [
{
"Id": 1,
"Name": "Adam Admin",
"Role": 1
},
{
"Id": 2,
"Name": "Emily Employee",
"Role": 4
},
{
"Id": 3,
"Name": "Sam Supervisor",
"Role": 3
},
{
"Id": 4,
"Name": "Mary Manager",
"Role": 2
},
{
"Id": 5,
"Name": "Steve Trainer",
"Role": 5
}
]
```

 - I have gone for the quickest solution. Here I've collected all the children roles of a specific role and then check the list of all users. If they belong to the children role then we print them out. 
 - I have also taken the roles and users objects as provided without any modifications to them

```python
def getchildrenroles(roleid, children=[]): # to get sub roles to current role
    for role in roles: 
        if role['Parent']==roleid: 
            children.append(role['Id'])
            getchildrenroles(role['Id'], children) 
    return children
```

```python
def getSubOrdinates(userid,childrenroles=[]): # to get subordinates to current user

    childrenroles=[getchildrenroles(user['Role'],[]) for user in users if user['Id']==userid][0]
    children=[user for user in users if user['Id'] in childrenroles]
    if(len(children)==0):
        print('sorry no subordinates for given user')
    else: 
        [print(child) for child in children]
    return children
```
